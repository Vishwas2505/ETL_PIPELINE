```python
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/WA_Fn-UseC_-Telco-Customer-Churn.csv')
df.head()
```

```python
#cleaning of data
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")
```

```python
df["tenure"] = df["tenure"].fillna(df["tenure"].median())
df["MonthlyCharges"] = df["MonthlyCharges"].fillna(df["MonthlyCharges"].median())
df["TotalCharges"] = df["TotalCharges"].fillna(df["TotalCharges"].median())
```

```python
cat_cols = df.select_dtypes(include="object").columns
df[cat_cols] = df[cat_cols].fillna("Unknown")

df.head()
```

```python
df.isnull().sum()
```

```python
#Feature Engineering

# 1. tenure_group
df["tenure_group"] = pd.cut(
    df["tenure"],
    bins=[0, 12, 36, 60, 1000],
    labels=["New", "Regular", "Loyal", "Champion"]
)
```

```python
df.head()
```

```python
# 2. monthly_charge_segment
df["monthly_charge_segment"] = pd.cut(
    df["MonthlyCharges"],
    bins=[0, 30, 70, 1000],
    labels=["Low", "Medium", "High"],
    right=False
)
```

```python
df.head()
```

```python
# ✔ . has_internet_service
df["has_internet_service"] = df["InternetService"].map({
    "DSL": 1,
    "Fiber optic": 1,
    "No": 0
}).fillna(0)
```

```python
df.head()
```

```python
df
```

```python
#  4. is_multi_line_user
df["is_multi_line_user"] = df["MultipleLines"].apply(
    lambda x: 1 if x == "Yes" else 0
)
df.head()
```

```python
#  5. contract_type_code
df["contract_type_code"] = df["Contract"].map({
    "Month-to-month": 0,
    "One year": 1,
    "Two year": 2
}).fillna(0)

df.head()
```

```python
# Drop unnecessary fields

df = df.drop(columns=["customerID", "gender"], errors="ignore")
df.to_csv("telco_cleaned.csv",index=False)
df.head()
```
